```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

class InstagramScrapper:
    # Fonction pour r√©cup√©rer les informations d'un profil Instagram
    def getinfo(self, url):
        # Envoie une requ√™te GET √† l'URL
        response = requests.get('https://' + url)
        
        # R√©cup√®re le contenu HTML de la page
        html = response.text
        
        # Parse le contenu HTML
        soup = BeautifulSoup(html, 'html.parser')
        
        # Trouve tous les tags 'meta' avec l'attribut 'property' √©gal √† 'og:description'
        data = soup.find_all('meta', attrs={'property': 'og:description'})
        
        # Si il y a au moins un tel tag
        if len(data) > 0:
            # R√©cup√®re le contenu du tag et le split en une liste de mots
            text = data[0].get('content').split()
            
            # Le nom d'utilisateur commence apr√®s le mot 'from'
            start_index = text.index('from') + 1
            
            # Le nom d'utilisateur se termine par le dernier mot de la liste
            at_name = text[-1].strip('()')
            
            # Le nom de l'utilisateur est les mots restants, s'il y en a
            user = " ".join(text[start_index:-1]) if len(text[start_index:-1]) > 0 else at_name.strip('@()')
            
            # Le premier mot est le nombre de followers
            followers = text[0]
            
            # Le troisi√®me mot est le nombre de comptes que l'utilisateur suit
            following = text[2]
            
            # Le cinqui√®me mot est le nombre de publications de l'utilisateur
            posts = text[4]
        else:
            # Si il n'y a pas de tel tag, le profil a √©t√© supprim√© ou bloqu√©
            user = at_name = followers = following = posts = ''
        
        # Renvoie le nom de l'utilisateur, son nom d'utilisateur, son nombre de followers, son nombre de comptes suivis, et son nombre de publications
        return (user, at_name, following, followers, posts)

    def main(self):
        # Ouvre le fichier 'starngage.txt'
        with open('ig_users.txt') as f:
            # Lit les lignes du fichier et les stocke dans une liste
            content = f.readlines()
            
            # Supprime les espaces en d√©but et fin de chaque ligne
            content = [x.strip() for x in content]
            
            # Cr√©e une liste vide pour stocker les tuples d'informations
            info_list = []
            
            # Parcours les URLs de la liste
            for url in content:
                # Obtient les informations pour chaque URL et ajoute-les √† la liste
                info_list.append(self.getinfo(url))
        
        # Cr√©er une trame de donn√©es √† partir de la liste des tuples d'informations
        df = pd.DataFrame(info_list, columns=['name', 'username', 'Following', 'Followers', 'Posts'])
        
        # Enregistrer DataFrame en un fichier CSV 
        df.to_csv('ig_users_data.csv', index=False)

if __name__ == '__main__':
    # Cr√©er une instance de la classe InstagramScrapper
    obj = InstagramScrapper()
    obj.main()

        
```
